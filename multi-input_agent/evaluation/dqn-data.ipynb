```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
def plot_by_step(df, title):
    plt.plot(df["Step"], df["Value"])
    plt.xlabel("Step")
    plt.ylabel("Value")
    plt.title(title)


def plot_by_time(df,title):
    t0 = df["Wall time"][0]
    time_ax = []
    for t in df["Wall time"]:
        time_ax.append(t- t0)
    plt.plot(time_ax, df["Value"])
    plt.xlabel("Time (seconds)")
    plt.ylabel("Value")
    plt.title(title)
```

```python

df_return = pd.read_csv(r".\DQN\multi-input_agent-results\episodic_return.csv")
df_loss = pd.read_csv(r".\DQN\multi-input_agent-results\total_loss.csv")
```

## DQN plots by steps

```python
plot_by_step(df_return, "Episodic return")
```

```python
plot_by_step(df_loss, "Total loss")
```

## DQN plots by time

```python
plot_by_time(df_return, "Episodic return")
```

```python
plot_by_time(df_loss, "Total loss")
```
